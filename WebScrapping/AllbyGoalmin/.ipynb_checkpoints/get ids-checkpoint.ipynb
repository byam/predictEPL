```python
import bs4
import requests
import os
import sys
```

```python
xscores_url = "http://www.xscores.com/soccer/leagueresults/england/premier_league/2014-2015/r/"
round_id = 125645
all_rounds = 38
```

```python
def get_ids(url, round):
    res = requests.get(url + str(round))
    if res.status_code != 200:
        sys.exit()
    soup = bs4.BeautifulSoup(res.text, "html.parser")
    scorers_class = soup.findAll('span', {'class': 'scorers'})
    rpc_ids = list(map(lambda scorer: int(str(scorer).split('|')[1]), scorers_class))
    return rpc_ids
```

```python
dic_round_rcps = {}
for i in range(all_rounds):
    dic_round_rcps[i+1] = get_ids(xscores_url, round_id + i)
```
