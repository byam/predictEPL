```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
import codecs
import collections
```

### Define

```python
home = "NUFC"
away = "Arsenal"
```

### Load Data and Preprocessing

```python
# Read file
tweets_file = open("1508292045_NUFCvsARSENAL.txt", "r")
```

```python
# Parsing JSON

tweets_data = []
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
```

```python
tweets_file.close()
```

```python
print len(tweets_data)
```

```python
# tweets_data[100]
```

### Create DataFrame as tweets

```python
tweets = pd.DataFrame()
tweets['text'] = map(lambda tweet: tweet['text'], tweets_data)
tweets['lang'] = map(lambda tweet: tweet['lang'], tweets_data)
tweets['country'] = map(lambda tweet: tweet['place']['country'] if tweet['place'] != None else None, tweets_data)
tweets['hashtags'] = map(lambda tweet: map(lambda tag: tag['text'], tweet['entities']['hashtags']), tweets_data)
```

```python
tweets[home] = map(lambda tags: home in tags, tweets['hashtags'])
tweets[away] = map(lambda tags: away in tags, tweets['hashtags'])
```

```python
tweets.head()
```

```python
print "Home Team Hashtags: ", len(tweets[tweets[home] == True])
print "Away Team Hashtags: ", len(tweets[tweets[away] == True])
```

```python
slices = [7,2,2,13]
activities = ['home','away','both']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow= True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.title('Interesting Graph\nCheck it out')
plt.show()
```

### Top 5 languages

```python
tweets_by_lang = tweets['lang'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Languages', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 languages', fontsize=15, fontweight='bold')
tweets_by_lang[:5].plot(ax=ax, kind='bar', color='red')
tweets_by_lang[:5]
```

### Top 5 countries

```python
tweets_by_country = tweets['country'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Countries', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 countries', fontsize=15, fontweight='bold')
tweets_by_country[:5].plot(ax=ax, kind='bar', color='blue')
tweets_by_country[:5]
```

### Count All Words

```python
tweets_eng = tweets[tweets["lang"] == "en"]
tweets_eng.head()
```

```python
print len(tweets_eng)
```

```python
# create all words lists
all_words = []

for tweet in tweets_eng["text"]:
    tweet_words = word_tokenize(tweet)
    for word in tweet_words:
        all_words.append(word.lower())
```

```python
# count all words
counted_all_words = collections.Counter(all_words)
counted_all_words.most_common()[:20]
```

### Extract Adjective Words

```python
tweets_eng['text'][16794]
```

```python
nltk.pos_tag(word_tokenize(tweets_eng['text'][16794]))
```

```python
# create J all words lists
all_J_words = []

#  j is adject, r is adverb, and v is verb
#allowed_word_types = ["J","R","V"]
allowed_word_types = ["J"]

for tweet in tweets_eng["text"]:
    
    tweet_words = word_tokenize(tweet)
    
    tweet_words_pos = nltk.pos_tag(tweet_words)
    
    for word in tweet_words_pos:
        if word[1][0] in allowed_word_types and word[0][0] != "/":
            all_J_words.append(word[0].lower())
```

```python
# count J all words
counted_all_J_words = collections.Counter(all_J_words)
counted_all_J_words.most_common()[:20]
```
