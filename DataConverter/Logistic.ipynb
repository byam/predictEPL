```python
def create_minute_dics_sentiment(df):
    minutes = range(120)
    
    dic_minutes_home = {}
    dic_minutes_away = {}
    dic_minutes_both = {}
    
    dic_minutes_home_cum = {}
    dic_minutes_away_cum = {}
    dic_minutes_both_cum = {}
    
    dic_minutes_home_pos = {}
    dic_minutes_home_neg = {}
    dic_minutes_away_pos = {}
    dic_minutes_away_neg = {}
    
    dic_minutes_home_pos_cum = {}
    dic_minutes_home_neg_cum = {}
    dic_minutes_away_pos_cum = {}
    dic_minutes_away_neg_cum = {}
    
    dic_minutes_home_pos_cum_score = {}
    dic_minutes_home_neg_cum_score = {}
    dic_minutes_away_pos_cum_score = {}
    dic_minutes_away_neg_cum_score = {}

    for minute in minutes:
        # home-away-both
        try:
            dic_minutes_home[minute] = len(df[(df['side'] == 'home') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_home[minute] = 0


        try:
            dic_minutes_away[minute] = len(df[(df['side'] == 'away') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_away[minute] = 0


        try:
            dic_minutes_both[minute] = len(df[(df['side'] == 'both') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_both[minute] = 0
            
        
        # home-away-both-cum
        try:
            dic_minutes_home_cum[minute] = len(df[(df['side'] == 'home') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_home_cum[minute] = 0


        try:
            dic_minutes_away_cum[minute] = len(df[(df['side'] == 'away') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_away_cum[minute] = 0


        try:
            dic_minutes_both_cum[minute] = len(df[(df['side'] == 'both') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_both_cum[minute] = 0
        
    
        
        # home
        try:
            dic_minutes_home_pos[minute] = len(df[(df['side'] == 'home') & \
                                              (df['long_sent'] == 'pos') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_home_pos[minute] = 0
        
        try:
            dic_minutes_home_neg[minute] = len(df[(df['side'] == 'home') & \
                                              (df['long_sent'] == 'neg') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_home_neg[minute] = 0
        
        # home-cum
        try:
            dic_minutes_home_pos_cum[minute] = len(df[(df['side'] == 'home') & \
                                              (df['long_sent'] == 'pos') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_home_pos_cum[minute] = 0
        
        try:
            dic_minutes_home_neg_cum[minute] = len(df[(df['side'] == 'home') & \
                                              (df['long_sent'] == 'neg') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_home_neg_cum[minute] = 0
        
        # away
        try:
            dic_minutes_away_pos[minute] = len(df[(df['side'] == 'away') & \
                                              (df['long_sent'] == 'pos') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_away_pos[minute] = 0
        
        try:
            dic_minutes_away_neg[minute] = len(df[(df['side'] == 'away') & \
                                              (df['long_sent'] == 'neg') & \
                                              (df['ith_minute'] > minute) & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_away_neg[minute] = 0
        
        # away-cum
        try:
            dic_minutes_away_pos_cum[minute] = len(df[(df['side'] == 'away') & \
                                              (df['long_sent'] == 'pos') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_away_pos_cum[minute] = 0
        
        try:
            dic_minutes_away_neg_cum[minute] = len(df[(df['side'] == 'away') & \
                                              (df['long_sent'] == 'neg') & \
                                              (df['ith_minute'] < minute + 1)])
        except:
            dic_minutes_away_neg_cum[minute] = 0
        
        # home-cum-score
        try:
            l = list(df[(df['side'] == 'home') & (df['ith_minute'] < minute + 1)]['long_pos'])
            dic_minutes_home_pos_cum_score[minute] = sum([float(x) for x in l])
        
        except:
            dic_minutes_home_pos_cum_score[minute] = 0
        
        try:
            l = list(df[(df['side'] == 'home') & (df['ith_minute'] < minute + 1)]['long_neg'])
            dic_minutes_home_neg_cum_score[minute] = sum([float(x) for x in l])
        
        except:
            dic_minutes_home_neg_cum_score[minute] = 0
        
        # away-cum-score
        try:
            l = list(df[(df['side'] == 'away') & (df['ith_minute'] < minute + 1)]['long_pos'])
            dic_minutes_away_pos_cum_score[minute] = sum([float(x) for x in l])
        
        except:
            dic_minutes_away_pos_cum_score[minute] = 0
        
        try:
            l = list(df[(df['side'] == 'away') & (df['ith_minute'] < minute + 1)]['long_neg'])
            dic_minutes_away_neg_cum_score[minute] = sum([float(x) for x in l])
        
        except:
            dic_minutes_away_neg_cum_score[minute] = 0
        
    return (dic_minutes_home, dic_minutes_away, dic_minutes_both,
            dic_minutes_home_cum, dic_minutes_away_cum, dic_minutes_both_cum,
            dic_minutes_home_pos, dic_minutes_home_neg,
            dic_minutes_away_pos, dic_minutes_away_neg,
            dic_minutes_home_pos_cum, dic_minutes_home_neg_cum,
            dic_minutes_away_pos_cum, dic_minutes_away_neg_cum,
            dic_minutes_home_pos_cum_score, dic_minutes_home_neg_cum_score,
            dic_minutes_away_pos_cum_score, dic_minutes_away_neg_cum_score)
```
