```python
%pylab inline
%matplotlib inline

import os
import csv
import pandas as pd
```

```python
# Return pos & neg's '%' dataframe
def dfSentPercent(dfSent):
    dfPer = dfSent.copy()
    dfPer['home_pos %'] = dfPer['home_pos'] / (dfPer['home_pos'] + dfPer['home_neg'])
    dfPer['home_neg %'] = dfPer['home_neg'] / (dfPer['home_pos'] + dfPer['home_neg'])
    dfPer['away_pos %'] = dfPer['away_pos'] / (dfPer['away_pos'] + dfPer['away_neg'])
    dfPer['away_neg %'] = dfPer['away_neg'] / (dfPer['away_pos'] + dfPer['away_neg'])
    return dfPer.iloc[:,[0,1,6,7,8,9]]
```

```python
print("Sentiment Result by %")
dfSentPercent(dfSentGW4)
```
