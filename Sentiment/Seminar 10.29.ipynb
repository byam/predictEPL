```python
def preprocess(x):
    # 不要な行, 列のフィルタなど、データサイズを削減する処理
    return x

def readData(home_team, away_team, GW):
    filePath = '/Users/Bya/Dropbox/Research/datas/Results/NB_SWN/' + GW +'/' + home_team + '_vs_' + away_team + '.csv'
    
    reader = pd.read_csv(filePath, chunksize=1000)
    dfHA = pd.concat((preprocess(r) for r in reader), ignore_index=True)
    
    return dfHA
```

```python
def res_nb_swn(home_team, away_team, GW):
    df = readData(home_team, away_team, GW)

    dfHome = df[(df['side'] == 'home')]
    dfAway = df[(df['side'] == 'away')]

    len_home = len(dfHome)
    len_away = len(dfAway)

    home_nb_pos_avg = sum(dfHome['nb_pos']) / len_home
    home_nb_neg_avg = sum(dfHome['nb_neg']) / len_home

    away_nb_pos_avg = sum(dfAway['nb_pos']) / len_away
    away_nb_neg_avg = sum(dfAway['nb_neg']) / len_away

    home_swn_pos_avg = sum(dfHome['swn_pos']) / len_home
    home_swn_neg_avg = sum(dfHome['swn_neg']) / len_home

    away_swn_pos_avg = sum(dfAway['swn_pos']) / len_away
    away_swn_neg_avg = sum(dfAway['swn_neg']) / len_away

    home_nb_pos = len(df[(df['nb'] == 'pos') & (df['side'] == 'home')]) / len_home
    home_nb_neg = len(df[(df['nb'] == 'neg') & (df['side'] == 'home')]) / len_home
    away_nb_pos = len(df[(df['nb'] == 'pos') & (df['side'] == 'away')]) / len_away
    away_nb_neg = len(df[(df['nb'] == 'neg') & (df['side'] == 'away')]) / len_away

    home_swn_pos = len(df[(df['swn'] == 'pos') & (df['side'] == 'home')]) / len_home
    home_swn_neg = len(df[(df['swn'] == 'neg') & (df['side'] == 'home')]) / len_home
    away_swn_pos = len(df[(df['swn'] == 'pos') & (df['side'] == 'away')]) / len_away
    away_swn_neg = len(df[(df['swn'] == 'neg') & (df['side'] == 'away')]) / len_away

    res = [
        home_team, away_team,
        home_nb_pos, home_nb_neg, away_nb_pos, away_nb_neg,
        home_swn_pos, home_swn_neg, away_swn_pos, away_swn_neg,
           ]

    filename = GW + '_nb_swn'
    filepathName = '/Users/Bya/Dropbox/Research/datas/Results/NB_SWN/'
    dataIO.ListSaveToCSVappend(res, filename, filepathName)
    
    
    res = [
        home_team, away_team,
        home_nb_pos_avg, home_nb_neg_avg, away_nb_pos_avg, away_nb_neg_avg,
        home_swn_pos_avg, home_swn_neg_avg, away_swn_pos_avg, away_swn_neg_avg,   
           ]

    filename = GW + '_nb_swn_average'
    filepathName = '/Users/Bya/Dropbox/Research/datas/Results/NB_SWN/'
    dataIO.ListSaveToCSVappend(res, filename, filepathName)
```

```python
team = Teams()
weekGames = [
    team.Newcastle, team.Arsenal,
    
    team.Villa, team.Sunderland,
    team.Bournemouth, team.Leicester,
    team.Chelsea, team.Crystal,
    team.Liverpool, team.WestHam,
    team.City, team.Watford,
    team.Stoke, team.WestBromwich,
    
#     team.Tottenham, team.Everton,
    team.Southampton, team.Norwich,
    team.Swansea, team.United,
            ]
weekGames = createTuples(weekGames)

header = [
        'home_team', 'away_team',
        'home_nb_pos', 'home_nb_neg', 'away_nb_pos', 'away_nb_neg',
        'home_swn_pos', 'home_swn_neg', 'away_swn_pos', 'away_swn_neg',
        ]

header_avg = [
        'home_team', 'away_team',
        'home_nb_pos_avg', 'home_nb_neg_avg', 'away_nb_pos_avg', 'away_nb_neg_avg',
        'home_swn_pos_avg', 'home_swn_neg_avg', 'away_swn_pos_avg', 'away_swn_neg_avg',   
        ]



GW = 'GW4'
filename = GW + '_nb_swn'
filepathName = '/Users/Bya/Dropbox/Research/datas/Results/NB_SWN/'
dataIO.ListSaveToCSVappend(header, filename, filepathName)

filename = GW + '_nb_swn_average'
filepathName = '/Users/Bya/Dropbox/Research/datas/Results/NB_SWN/'
dataIO.ListSaveToCSVappend(header_avg, filename, filepathName)

for weekGame in weekGames:
    home_team = weekGame[0]
    away_team = weekGame[1]
        
    res_nb_swn(home_team, away_team, GW)
```
